```python
from tmdbv3api import TMDb
import json
import requests
tmdb=TMDb()
tmdb.api_key='643b71da3b8e6b8deb2a4171b281bc7d'
```

```python
from tmdbv3api import Movie
tmdb_movie=Movie()
def get_genre(x):
    genres=[]
    result=tmdb_movie.search(x)
    movie_id=result[0].id
    #print(movie_id)
    response=requests.get('https://api.themoviedb.org/3/movie/{}?api_key=643b71da3b8e6b8deb2a4171b281bc7d'.format(movie_id))
    data_json=response.json()
    #print(data_json)
    if data_json['genres']:
        print(data_json['genres'])
        genre_str=" "
        for i in range(0,len(data_json['genres'])):
            genres.append(data_json['genres'][i]['name'])
        return genre_str.join(genres)
    else:
        np.NaN
```

```python
df_f['genres']=df_f['Title'].map(lambda x: get_genre(str(x)))
```

```python
df_2020=df_f[['Title','Cast and crew','genres']]
```

```python
df_2020
```

```python
def get_director(x):
    if " (director)" in x:
        return x.split(" (director)")[0]
    elif " (directors)" in x:
        return x.split(" (directors)")[0]
    else:
        return x.split(" (director/screenplay)")[0]
```

```python
df_2020['director_name']=df_2020['Cast and crew'].map(lambda x: get_director(str(x)))
```

```python
df_2020.dtypes
```

```python
def get_actor1(x):
    return ((x.split("screenplay); ")[-1]).split(", ")[0])
```

```python
df_2020['actor_1_name']=df_2020['Cast and crew'].map(lambda x: get_actor1(str(x)))
```

```python
def get_actor2(x):
    if len((x.split("screenplay); ")[-1]).split(", "))<2:
        return np.nan
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[1])
```

```python
 df_2020['actor_2_name']=df_2020['Cast and crew'].map(lambda x: get_actor2(str(x)))
```

```python
def get_actor3(x):
    if len((x.split("screenplay); ")[-1]).split(", "))<3:
        return np.nan
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[2])
    
    
```

```python
df_2020['actor_3_name']=df_2020['Cast and crew'].map(lambda x: get_actor3(str(x)))
```

```python
df_2020=df_2020.rename(columns={'Title':'movie_title'})
```

```python
new_df20=df_2020.loc[:,['director_name',
       'actor_1_name', 'actor_2_name', 'actor_3_name','genres','movie_title']]
```
