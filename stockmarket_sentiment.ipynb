```python
df=pd.read_csv(r'C:\Users\pallaw\Desktop\ML&Deep_Learning\stockmarket_sentiment\Combined_News_DJIA.csv',encoding='ISO_8859-1')
```

```python
df
```

```python
train=df[df['Date']<'20150101']
test=df[df['Date']>'20150101']
```

```python
data=train.iloc[:,2:27]
data.replace("[^a-zA-Z]"," ",regex=True,inplace=True)
list1=[i for i in range(25)]
new_index=[str(i) for i in list1]
data.columns=new_index
data.head(5)
```

```python
for index in new_index:
    data[index]=data[index].str.lower()
data.head()    
```

```python
' '.join(str(x) for x in data.iloc[1,:25])
```

```python
data.index
```

```python
headlines=[]
for row in range(len(data.index)):
    headlines.append(' '.join(str(x) for x in data.iloc[row,:]))
```

```python
len(headlines)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
countvector=CountVectorizer(ngram_range=(2,2))
traindataset=countvector.fit_transform(headlines)
```

```python
randomclassfier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassfier.fit(traindataset,train['Label'])
```

```python
test_transform=[]
for row in range(len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,:]))
test_dataset=countvector.transform(test_transform)
predictions=randomclassfier.predict(test_dataset)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
predictions
```

```python
matrix=confusion_matrix(test['Label'],predictions)
print(matrix)
score=accuracy_score(test['Label'],predictions)
print(score)
report=classification_report(test['Label'],predictions)
print(report)
```
